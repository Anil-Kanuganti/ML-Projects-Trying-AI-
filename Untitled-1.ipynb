```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('train.csv')
sns.set_style('whitegrid')
print("dataloaded:",df.shape)
df.info()
df.describe()
df['Age'].hist()
sns.heatmap(df.isnull())
df.groupby('Survived')['Age'].mean()
df.shape
df.columns
df.dtypes
df.head(10)
df.isnull().sum()
df['Age'].fillna(df['Age'].sum())
df.dropna(subset=['Embarked'])
df[df['Survived']==1]
df[df['Age']>30]
df.sort_values('Fare', ascending=False).head(10)
df.groupby('Sex')['Survived'].mean()
df.groupby('Pclass')['Survived'].mean()
df.groupby('Pclass')['Fare'].mean() 
```

Survival Count
Died count is more than the survived count which is like more than 75%

```python
sns.countplot(x='Survived',data=df)
plt.title('survival count(0=died,1=Survived)')
plt.savefig('Survived Chart1.png')
plt.show()
```

```python
sns.countplot(x='Sex',hue='Survived',data=df)
plt.title('Survival by Gender')
plt.savefig('Chart2 Survival by Gender.png')
plt.show()
```

Survived count comparing by the gender
Compare to the male female are the more number of survivors and females has the less number of the deaths compare to men. Where as the ratio of female survivors is double than the male survivors. 

```python
df['Age'].hist(bins=30, color='steelblue',edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.savefig('Chart3 Distribution of age.png')
plt.show()
```

Distribution of Age
By absorbing the above graph I can see that more people in the Titanic is between 20-40 age group.and the second highest is looks like 0-3 years of age. 

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.xlabel('Class(1=First, 2=Second, 3=Third)')
plt.title('Survival Range by Passenger class')
plt.savefig('Chart4_class.png')
plt.show()
```

Passenger Class Survival
Comapring to the all 3 levels fo the passenger levels The Top survival is the First class and followed by second and Third class. 

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, alpha=0.6)
plt.title('Age VS Fare Survived')
plt.savefig('chart5_AgevsFare')
plt.show()
```

## Who survived?
Women survived at 74% rate vs men at only 19%.
First class passengers survived more than third class.
